```python
import torch

class TocoNet(torch.nn.Module):
    def __init__(self):
        super().__init__()
        # --- ENCODER ---
        self.conv1_1 = torch.nn.Conv1d(in_channels=2, out_channels=8, kernel_size=7, padding=3)
        self.bn1_1   = torch.nn.BatchNorm1d(num_features=8) 
        self.relu1_1 = torch.nn.ReLU()
        self.conv2_1 = torch.nn.Conv1d(in_channels=8, out_channels=8, kernel_size=7, padding=3)
        self.bn2_1   = torch.nn.BatchNorm1d(num_features=8)
        self.relu2_1 = torch.nn.ReLU()
        self.pool1   = torch.nn.MaxPool1d(kernel_size=2, stride=2)

        self.conv1_2 = torch.nn.Conv1d(in_channels=8, out_channels=16, kernel_size=7, padding=3)
        self.bn1_2   = torch.nn.BatchNorm1d(num_features=16) 
        self.relu1_2 = torch.nn.ReLU()
        self.conv2_2 = torch.nn.Conv1d(in_channels=16, out_channels=16, kernel_size=7, padding=3)
        self.bn2_2   = torch.nn.BatchNorm1d(num_features=16) 
        self.relu2_2 = torch.nn.ReLU()
        self.pool2   = torch.nn.MaxPool1d(kernel_size=2, stride=2)

        self.conv1_3 = torch.nn.Conv1d(in_channels=16, out_channels=64, kernel_size=7, stride=4, padding=3)
        self.bn1_3   = torch.nn.BatchNorm1d(num_features=64) 
        self.relu1_3 = torch.nn.ReLU()
        self.conv2_3 = torch.nn.Conv1d(in_channels=64, out_channels=128, kernel_size=7, padding=3)
        self.bn2_3   = torch.nn.BatchNorm1d(num_features=128) 
        self.relu2_3 = torch.nn.ReLU()
        self.conv3_3 = torch.nn.Conv1d(in_channels=128, out_channels=128, kernel_size=7, padding=3)
        self.bn3_3   = torch.nn.BatchNorm1d(num_features=128) 
        self.relu3_3 = torch.nn.ReLU()
        self.pool3   = torch.nn.MaxPool1d(kernel_size=2, stride=2)
        
        # --- BOTTLENECK ---
        self.conv1_4 = torch.nn.Conv1d(in_channels=128, out_channels=256, kernel_size=7, padding=3)
        self.bn1_4   = torch.nn.BatchNorm1d(num_features=256) 
        self.relu1_4 = torch.nn.ReLU()
        self.conv2_4 = torch.nn.Conv1d(in_channels=256, out_channels=256, kernel_size=7, padding=3)
        self.bn2_4   = torch.nn.BatchNorm1d(num_features=256) 
        self.relu2_4 = torch.nn.ReLU()
        self.conv3_4 = torch.nn.Conv1d(in_channels=256, out_channels=256, kernel_size=7, padding=3)
        self.bn3_4   = torch.nn.BatchNorm1d(num_features=256) 
        self.relu3_4 = torch.nn.ReLU()
        self.up1     = torch.nn.ConvTranspose1d(in_channels=256, out_channels=256, kernel_size=2, stride=2)

        # --- DECODER ---
        self.conv1_5 = torch.nn.Conv1d(in_channels=256 + 128, out_channels=128, kernel_size=7, padding=3)
        self.bn1_5   = torch.nn.BatchNorm1d(num_features=128) 
        self.relu1_5 = torch.nn.ReLU()
        self.conv2_5 = torch.nn.Conv1d(in_channels=128, out_channels=128, kernel_size=7, padding=3)
        self.bn2_5   = torch.nn.BatchNorm1d(num_features=128)
        self.relu2_5 = torch.nn.ReLU()
        self.up2     = torch.nn.ConvTranspose1d(in_channels=128, out_channels=32, kernel_size=8, stride=8) 

        self.conv1_6 = torch.nn.Conv1d(in_channels=32 + 16, out_channels=16, kernel_size=7, padding=3)
        self.bn1_6   = torch.nn.BatchNorm1d(num_features=16) 
        self.relu1_6 = torch.nn.ReLU()
        self.conv2_6 = torch.nn.Conv1d(in_channels=16, out_channels=16, kernel_size=7, padding=3)
        self.bn2_6   = torch.nn.BatchNorm1d(num_features=16)
        self.relu2_6 = torch.nn.ReLU()
        self.up3     = torch.nn.ConvTranspose1d(in_channels=16, out_channels=16, kernel_size=2, stride=2) 

        self.conv1_7 = torch.nn.Conv1d(in_channels=16 + 8, out_channels=8, kernel_size=7, padding=3)
        self.bn1_7   = torch.nn.BatchNorm1d(num_features=8) 
        self.relu1_7 = torch.nn.ReLU()
        self.conv2_7 = torch.nn.Conv1d(in_channels=8, out_channels=8, kernel_size=7, padding=3)
        self.bn2_7   = torch.nn.BatchNorm1d(num_features=8) 
        self.relu2_7 = torch.nn.ReLU()

        self.conv3_7 = torch.nn.Conv1d(in_channels=8, out_channels=1, kernel_size=7, padding=3)
        self.bn3_7   = torch.nn.BatchNorm1d(num_features=1) 
        
        # REMOVED: self.sigmoid 

    def forward(self, x):
        x1 = self.relu2_1(self.bn2_1(self.conv2_1(self.relu1_1(self.bn1_1(self.conv1_1(x))))))
        p1 = self.pool1(x1)
        
        x2 = self.relu2_2(self.bn2_2(self.conv2_2(self.relu1_2(self.bn1_2(self.conv1_2(p1))))))
        p2 = self.pool2(x2)
        
        x3 = self.relu1_3(self.bn1_3(self.conv1_3(p2)))
        x3 = self.relu2_3(self.bn2_3(self.conv2_3(x3)))
        x3 = self.relu3_3(self.bn3_3(self.conv3_3(x3)))
        p3 = self.pool3(x3)
        
        x4 = self.relu1_4(self.bn1_4(self.conv1_4(p3)))
        x4 = self.relu2_4(self.bn2_4(self.conv2_4(x4)))
        x4 = self.relu3_4(self.bn3_4(self.conv3_4(x4)))
        u1 = self.up1(x4)
        
        u1 = torch.cat([u1, x3], dim=1) 
        x5 = self.relu2_5(self.bn2_5(self.conv2_5(self.relu1_5(self.bn1_5(self.conv1_5(u1))))))
        u2 = self.up2(x5)
        
        u2 = torch.cat([u2, x2], dim=1)
        x6 = self.relu2_6(self.bn2_6(self.conv2_6(self.relu1_6(self.bn1_6(self.conv1_6(u2))))))
        u3 = self.up3(x6)
        
        u3 = torch.cat([u3, x1], dim=1)
        x7 = self.relu2_7(self.bn2_7(self.conv2_7(self.relu1_7(self.bn1_7(self.conv1_7(u3))))))
        
        # Output raw logits instead of probabilities
        return self.bn3_7(self.conv3_7(x7))
```

```python
import json
import torch
import numpy as np
from torch.utils.data import Dataset
from scipy.signal import butter, lfilter, filtfilt

def butter_lowpass_filter(data, cutoff=0.05, fs=4.0, order=2):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return filtfilt(b, a, data)

class GlobalCTGDataset(Dataset):
    def __init__(self, chunks_list, sample_fs=4.0):
        """
        Instead of loading files directly, this takes a pre-split list of chunks.
        """
        self.chunks = chunks_list
        self.fs = sample_fs
        self.target_length = 4800  # Strict U-Net target timeline size

    def __len__(self):
        return len(self.chunks)

    def __getitem__(self, idx):
        chunk = self.chunks[idx]
        raw_signal = np.array(chunk["signal_values"], dtype=np.float32)
        
        # --- FIX: Pad or Truncate raw_signal to exactly 4800 elements ---
        current_length = len(raw_signal)
        if current_length > self.target_length:
            raw_signal = raw_signal[:self.target_length]
        elif current_length < self.target_length:
            padding_needed = self.target_length - current_length
            raw_signal = np.pad(raw_signal, (0, padding_needed), mode='constant', constant_values=0)
        
        #Low pass filter (Now consistently processing exactly 4800 elements)
        filtered_signal = butter_lowpass_filter(raw_signal, cutoff=0.05, fs=self.fs, order=2).astype(np.float32)
        
        #Scale directly by clinical bounds [0, 127 mmHg]
        input_tensor = torch.tensor(np.stack([raw_signal / 127.5, filtered_signal / 127.5], axis=0), dtype=torch.float32)
        
        #Create a strict 4800 length label mask from 'uc_events'
        label_mask = np.zeros(self.target_length, dtype=np.float32)
        if "contractions" in chunk:
            for event in chunk["contractions"]:
                start_idx = event.get("start_idx", 0)
                end_idx = event.get("end_idx", current_length)
                
                # Enforce limits so index slices never overshoot our 4800 tensor cap
                start_idx = min(max(0, start_idx), self.target_length)
                end_idx = min(max(0, end_idx), self.target_length)
                
                label_mask[start_idx:end_idx] = 1.0
                
        label_tensor = torch.tensor(label_mask, dtype=torch.float32).unsqueeze(0) # Shape: [1, 4800]
        
        return input_tensor, label_tensor
```

```python
import os
import glob
import json
from sklearn.model_selection import train_test_split
from torch.utils.data import DataLoader

json_folder_path = "json_data"
json_files = glob.glob(os.path.join(json_folder_path, "*.json"))

if not json_files:
    print(f"Error: No JSON files found in '{json_folder_path}'. Please check your folder structure.")

# Read and pool all chunks together dynamically
all_combined_chunks = []
for file_path in json_files:
    with open(file_path, 'r') as f:
        file_data = json.load(f)
        if isinstance(file_data, dict):
            all_combined_chunks.extend(list(file_data.values()))
        else:
            all_combined_chunks.extend(file_data)

print(f"\nTotal chunks gathered across all {len(json_files)} found scenarios: {len(all_combined_chunks)}")

# Shuffle and Split
train_chunks, test_chunks = train_test_split(
    all_combined_chunks, 
    test_size=0.20, 
    shuffle=True, 
    random_state=42
)

print(f"Training chunks: {len(train_chunks)}")
print(f"Testing chunks:  {len(test_chunks)}")

train_dataset = GlobalCTGDataset(chunks_list=train_chunks)
test_dataset = GlobalCTGDataset(chunks_list=test_chunks)


train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=1, shuffle=False)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch


def plot_test_sample(model, test_dataset, sample_idx=0):
    device = torch.device(
        "mps"
        if torch.backends.mps.is_available()
        else ("cuda" if torch.cuda.is_available() else "cpu")
    )
    model.eval()

    inputs, target = test_dataset[sample_idx]

    # Add batch dimension and push to device (MPS/CUDA/CPU)
    inputs_batch = inputs.unsqueeze(0).to(device)

    with torch.no_grad():
        output = model(inputs_batch)
        pred = (output > 0.5).float().cpu().squeeze().numpy()

    raw_signal = inputs[0].numpy()
    actual_mask = target.squeeze().numpy()

    # --- STACKED PLOTS CONFIGURATION ---
    # We create a figure with 3 rows stacked vertically, sharing the same X-axis (time-steps)
    fig, axs = plt.subplots(3, 1, figsize=(15, 10), sharex=True)

    # 1. Top Subplot: Raw Signal Data
    axs[0].plot(raw_signal, label="Normalized Raw CTG (Toco)", color="gray")
    axs[0].set_title(
        f"Model Evaluation vs Ground Truth (Test Sample {sample_idx})"
    )
    axs[0].set_ylabel("Signal Amplitude")
    axs[0].grid(True, alpha=0.3)
    axs[0].legend(loc="upper right")

    # 2. Middle Subplot: Ground Truth Segmentation Mask
    axs[1].fill_between(
        range(len(actual_mask)),
        0,
        1,
        where=actual_mask == 1,
        color="green",
        alpha=0.3,
        label="Ground Truth UC",
    )
    axs[1].set_ylabel("GT Mask (0/1)")
    axs[1].set_ylim(-0.1, 1.1)
    axs[1].grid(True, alpha=0.3)
    axs[1].legend(loc="upper right")

    # 3. Bottom Subplot: Model Predictions
    axs[2].plot(pred, label="Model UC Prediction", color="red", linestyle="--")
    axs[2].set_xlabel("Time-step index")
    axs[2].set_ylabel("Pred Mask (0/1)")
    axs[2].set_ylim(-0.1, 1.1)
    axs[2].grid(True, alpha=0.3)
    axs[2].legend(loc="upper right")

    # Clean up layout pacing to avoid overlapping axis labels
    plt.tight_layout()
    plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import torch
from sklearn.metrics import f1_score, precision_score, recall_score


def evaluate_and_plot_model(model, test_loader, max_plots=5):
    device = torch.device(
        "mps"
        if torch.backends.mps.is_available()
        else ("cuda" if torch.cuda.is_available() else "cpu")
    )
    model.eval()

    all_predictions = []
    all_ground_truths = []

    fig, axs = plt.subplots(
        max_plots, 1, figsize=(15, 4 * max_plots), squeeze=False
    )
    plot_counter = 0

    with torch.no_grad():
        for inputs, targets in test_loader:
            inputs = inputs.to(device)
            outputs = model(inputs)

            # Threshold logits at 0.0
            preds = (outputs > 0.0).float()

            all_predictions.extend(preds.cpu().view(-1).numpy())
            all_ground_truths.extend(targets.view(-1).numpy())

            if plot_counter < max_plots:
                batch_size = inputs.size(0)
                for b in range(batch_size):
                    if plot_counter >= max_plots:
                        break

                    ax = axs[plot_counter, 0]

                    raw_signal = inputs[b, 0].cpu().numpy()
                    actual_mask = targets[b].cpu().squeeze().numpy()
                    pred_mask = preds[b].cpu().squeeze().numpy()

                    # 1. Plot the raw signal trace line
                    ax.plot(
                        raw_signal,
                        label="Normalized Raw CTG (Toco)",
                        color="gray",
                        alpha=0.6,
                    )

                    # 2. NEW: Detect Ground Truth division boundaries (transitions)
                    # Find indices where mask goes from 0 to 1 or 1 to 0
                    diff = np.diff(actual_mask, prepend=0, append=0)
                    start_indices = np.where(diff == 1)[0]
                    end_indices = np.where(diff == -1)[0] - 1

                    # Draw vertical lines for the boundaries of Ground Truth labels
                    for idx, (start, end) in enumerate(
                        zip(start_indices, end_indices)
                    ):
                        # Line marking the start of a contraction label
                        ax.axvline(
                            x=start,
                            color="green",
                            linestyle="--",
                            linewidth=1.8,
                            label="GT Start" if idx == 0 else "",
                        )
                        # Line marking the end of a contraction label
                        ax.axvline(
                            x=end,
                            color="darkgreen",
                            linestyle="-",
                            linewidth=1.8,
                            label="GT End" if idx == 0 else "",
                        )

                        # Tiny text label at the top to number the events clearly
                        y_max = ax.get_ylim()[1]
                        ax.text(
                            (start + end) / 2,
                            y_max * 0.9,
                            f"UC {idx+1}",
                            color="darkgreen",
                            weight="bold",
                            horizontalalignment="center",
                        )

                    # 3. OVERLAY: Trace the model prediction line directly over it
                    ax.plot(
                        pred_mask * np.max(raw_signal),
                        label="Model UC Prediction",
                        color="red",
                        linestyle=":",
                        linewidth=2,
                    )

                    # Labels and formatting per individual sample plot
                    ax.set_title(
                        f"Model Prediction vs Ground Truth Boundaries (Sample {plot_counter})"
                    )
                    ax.set_ylabel("Signal Amplitude")
                    ax.grid(True, alpha=0.3)
                    ax.legend(loc="upper right")

                    plot_counter += 1

    # Calculate global dataset metrics
    precision = precision_score(
        all_ground_truths, all_predictions, zero_division=0
    )
    recall = recall_score(all_ground_truths, all_predictions, zero_division=0)
    f1 = f1_score(all_ground_truths, all_predictions, zero_division=0)

    print("\n--- TEST PERFORMANCE RESULTS ---")
    print(f"Precision : {precision:.4f}")
    print(f"Recall    : {recall:.4f}")
    print(f"F1-Score  : {f1:.4f}")

    axs[-1, 0].set_xlabel("Time-step index")

    plt.tight_layout()
    plt.show()

    return all_ground_truths, all_predictions
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

def train_model(train_loader, epochs=20):
    device = torch.device('mps' if torch.backends.mps.is_available() else ('cuda' if torch.cuda.is_available() else 'cpu'))
    print(f"Training using: {device}")
    
    # Calculate Class Imbalance Weight across your training dataset
    total_elements = 0
    total_positive = 0
    for _, targets in train_loader:
        total_elements += targets.numel()
        total_positive += targets.sum().item()
        
    total_negative = total_elements - total_positive
    # Compute ratio of negatives to positives
    pos_weight_val = total_negative / (total_positive if total_positive > 0 else 1.0)
    print(f"Dynamic balance positive weight multiplier calculated: {pos_weight_val:.4f}")
    
    model = TocoNet().to(device)
    
    # FIX: Use BCEWithLogitsLoss along with the weight tensor
    criterion = nn.BCEWithLogitsLoss(pos_weight=torch.tensor([pos_weight_val]).to(device))
    optimizer = optim.Adam(model.parameters(), lr=1e-3)
    
    for epoch in range(epochs):
        model.train()
        running_loss = 0.0
        
        for inputs, targets in train_loader:
            inputs = inputs.to(device)   
            targets = targets.to(device) 
            #print(targets)
            
            optimizer.zero_grad()
            
            outputs = model(inputs)      
            loss = criterion(outputs, targets)
            
            loss.backward()              
            optimizer.step()             
            
            running_loss += loss.item() * inputs.size(0)
            
        epoch_loss = running_loss / len(train_loader.dataset)
        if (epoch + 1) % 5 == 0 or epoch == 0:
            print(f"Epoch {epoch+1}/{epochs} - Loss: {epoch_loss:.4f}")
        
    print("Training finished successfully!")
    return model
```
